```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
path="./data/"
from google.colab import files

df=pd.read_excel("inf9_excel.xlsx")
df

# Descargar el archivo usando Google Colab
from google.colab import files
files.download("inf9_excel.xlsx")
```

```python
Em=df["Em"]
Kc=df["Kc"]
go=df["go"]


df['Fechaofi'] = df['dia'].astype(str) + '/' + df['mes'].astype(str) + '/' + df['año'].astype(str)
Fecha=df["Fechaofi"]


```

```python
plt.figure(figsize=(15, 6))
#plt.plot(Fecha,Em, color="dodgerblue")
plt.plot(Fecha,go, color="limegreen")
#plt.plot(Fecha,Kc, color="m")
plt.xlabel('Fecha')
plt.ylabel('Almacenamiento(mm)')
plt.title('Gráfica de Fecha vs Em')
plt.grid(True)

# Rotar las etiquetas del eje X para que se vean mejor
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.show()



```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.plot(Fecha, Em, color='tab:blue', linestyle='-', label='Em')
ax1.plot(Fecha, go, color='tab:orange', linestyle='-', label='go')

plt.xticks(rotation=45)
ax1.set_xlabel('Fecha')
ax1.set_ylabel('Em/go(mm)')
ax1.tick_params(axis='y')


ax2 = ax1.twinx()
ax2.plot(Fecha, Kc, color='tab:red',  linestyle='-', label='Kc')
ax2.set_ylabel('Kc')
ax2.tick_params(axis='y')
fig.legend(loc='upper right', bbox_to_anchor=(1,0.9))



plt.title('Variación temporal de Em, go y Kc')

plt.grid(True)
plt.show()
```
