```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definir símbolos y funciones
t = sp.symbols('t')
y = sp.Function('y')

# Definir la ecuación diferencial
ecuacion_diferencial = sp.Eq(y(t).diff(t), y(t)-cos(t))

# Definir la condición inicial y(0) = 1/4
condicion_inicial = {y(0): 1/4}

# Resolver la ecuación diferencial con dsolve
solucion = sp.dsolve(ecuacion_diferencial, ics=condicion_inicial)

# Obtener la solución en forma funcional
solucion_funcional = solucion.rhs

# Convertir la solución funcional a una función numérica
y_numerica = sp.lambdify(t, solucion_funcional, 'numpy')

# Graficar la solución
t_valores = np.linspace(-3, 10, 10000)
y_valores = y_numerica(t_valores)

plt.plot(t_valores, y_valores, label='Solución')
plt.scatter([0], [1/4], color='red', label='Condición Inicial (t=0, y=1/4)')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir el sistema de ecuaciones diferenciales
def sistema(y, t):
    x, y = y
    dydt = [-2*x + y, 3*x - 4*y]
    return dydt

# Condiciones iniciales
y0 = [1, 2]

# Definir puntos de tiempo para la gráfica
t_vals = np.linspace(0, 5, 100)

# Resolver el sistema de ecuaciones diferenciales con odeint
sol = odeint(sistema, y0, t_vals)

# Graficar las soluciones
plt.plot(t_vals, sol[:, 0], label='x(t)')
plt.plot(t_vals, sol[:, 1], label='y(t)')
plt.xlabel('t')
plt.ylabel('Soluciones')
plt.title('Soluciones del Sistema de Ecuaciones Diferenciales con Condiciones Iniciales')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir el sistema de ecuaciones diferenciales no homogéneas
def sistema(y, t):
    x, y = y
    dydt = [-2*x + y + np.sin(t), 3*x - 4*y + np.exp(-t)]
    return dydt

# Condiciones iniciales
y0 = [1, 2]

# Definir puntos de tiempo para la integración
t_vals = np.linspace(0, 5, 100)

# Resolver el sistema de ecuaciones diferenciales no homogéneas con odeint
sol = odeint(sistema, y0, t_vals)

# Graficar las soluciones
plt.plot(t_vals, sol[:, 0], label='x(t)')
plt.plot(t_vals, sol[:, 1], label='y(t)')
plt.xlabel('t')
plt.ylabel('Soluciones')
plt.title('Soluciones del Sistema de Ecuaciones Diferenciales No Homogéneas')
plt.legend()
plt.show()
```

Diagrama de fases

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir el sistema de ecuaciones diferenciales
def sistema(y, t):
    x, y = y
    dydt = [-2*x**2 + y, 3*x - 4*y]
    return dydt

# Definir la malla en el espacio de fase
x_vals = np.linspace(-2, 2, 20)
y_vals = np.linspace(-2, 2, 20)

# Crear una cuadrícula de condiciones iniciales
X, Y = np.meshgrid(x_vals, y_vals)

# Calcular las derivadas en cada punto de la cuadrícula
U, V = np.zeros_like(X), np.zeros_like(Y)
NI, NJ = X.shape

for i in range(NI):
    for j in range(NJ):
        x = X[i, j]
        y = Y[i, j]
        derivadas = sistema([x, y], 0)
        U[i,j] = derivadas[0]
        V[i,j] = derivadas[1]

# Normalizar las longitudes de los vectores para mejorar la visualización
norm = np.sqrt(U**2 + V**2)
U /= norm
V /= norm

# Graficar el diagrama de fases
plt.quiver(X, Y, U, V, scale=10, width=0.005)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Diagrama de Fases del Sistema de Ecuaciones Diferenciales')
plt.show()
```

```python
def dF(r, theta):
    return 0.5*r*(1 - r*r), 1+0*theta

X, Y = np.meshgrid(np.linspace(-3.0, 3.0, 30), np.linspace(-3.0, 3.0, 30))
R, Theta = (X**2 + Y**2)**0.5, np.arctan2(Y, X)
dR, dTheta = dF(R, Theta)
C, S = np.cos(Theta), np.sin(Theta)
U, V = dR*C - R*S*dTheta, dR*S+R*C*dTheta

plt.streamplot(X, Y, U, V, color='r', linewidth=0.5, density=1.6)
plt.axis('square')
plt.axis([-3, 3, -3, 3])
plt.show()
```

```python
from sympy import symbols, Eq, solve

# Definir las variables simbólicas
x, y = symbols('x y')

# Definir las ecuaciones diferenciales del sistema
xdot = x - y**2
ydot = y * (9*x - 4)

# Encontrar los puntos de equilibrio (donde las derivadas son cero)
equilibrio_x = solve(xdot, x)
equilibrio_y = solve(ydot, y)

# Mostrar los puntos de equilibrio
print("Puntos de equilibrio:")
print(f"x = {equilibrio_x}")
print(f"y = {equilibrio_y}")
```
